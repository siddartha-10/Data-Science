```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("glass.data")
```

```python
df
```

```python
# check for null values
df.isna().sum()
```

```python
data = df.drop(columns=['index','Class'],axis=1)
```

```python
data
```

```python
data.describe()
```

### we can see in the above table that the data is not standarized hence use standarization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_data = sc.fit_transform(data)
```

```python
df1 = pd.DataFrame(data=scaled_data,columns=data.columns)
```

```python
df1
```

```python
df1.describe()
```

### The above data looks standarized since mean is very close to zero

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
principlecomponent = pca.fit_transform(df1)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca = PCA(n_components=4)
new_data = pca.fit_transform(df1)
principal_df = pd.DataFrame(data=new_data,columns=[['principle component 1','principle component 2',
                                                    'principle component 3','principle component 4']])
```
