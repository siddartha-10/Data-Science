```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("glass.data")
```

```python
# check for null values
df.isna().sum()
```

```python
data = df.drop(columns=['index','Class'],axis=1)
```

```python
data.describe()
```

### we can see in the above table that the data is not standarized hence use standarization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_data = sc.fit_transform(data)
```

```python
df1 = pd.DataFrame(data=scaled_data,columns=data.columns)
```

```python
df1
```

```python
df1.describe()
```

### The above data looks standarized since mean is very close to zero

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
principlecomponent = pca.fit_transform(df1)
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
pca = PCA(n_components=4)
new_data = pca.fit_transform(df1)
principal_df = pd.DataFrame(data=new_data,columns=[['principle component 1','principle component 2',
                                                    'principle component 3','principle component 4']])
```

#### Application of PCA to digit data

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
digits
```

```python
X = digits.data
y = digits.target
```

```python
pca = PCA(2)
xproj = pca.fit_transform(X)
```

```python
print(X.shape)
print(xproj.shape)
```

```python
plt.scatter(xproj[:, 0], xproj[:, 1], c=y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar();
```

```python
sns.set()
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

**Pros of PCA:**

- Correlated features are removed.
- Model training time is reduced.
- Overfitting is reduced.
- Helps in better visualizations
- Ability to handle noise

**Cons of PCA**
- The resultant principal components are less interpretable than the original data
- Can lead to information loss if the explained variance threshold is not considered appropriately.
