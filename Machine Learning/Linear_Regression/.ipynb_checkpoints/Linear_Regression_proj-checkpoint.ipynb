```python
import pandas as pd 
import pickle
import numpy as np 
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def adj_r2(x,y):
    r2 = lm.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))
    return adjusted_r2
```

```python
data = pd.read_csv('Admission_Prediction.csv')
data.head()
```

```python
data.describe(include='all')
```

```python
data.info()
```

```python
## we can see from the above information that there are some missing values in the GRE Score, TOEFL Score column 
## and university ranking so we will fill them
data['GRE Score'] = data['GRE Score'].fillna(data['GRE Score'].mean())
data['TOEFL Score'] = data['TOEFL Score'].fillna(data['TOEFL Score'].mean())
data['University Rating'] = data['University Rating'].fillna(data['University Rating'].mode()[0])
# here we are using mode[0] because there maybe multiple values of the same frequency
```

```python
data.describe()
```

```python
data.info()
```

Now we can clearly see that there are no null values.

```python
# since we dont need the serial number we will drop it from our dataset
data = data.drop(columns=['Serial No.'])
data.head()
```

```python
# here we are seeing the distribution of each feature
plt.figure(figsize=(20,25),facecolor='white')
plotnumber = 1

for column in data:
    if plotnumber <= 16:
        ax = plt.subplot(4,4,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
# let's now see the relationship between the dependent and the independent features
y = data['Chance of Admit']
X = data.drop(columns = ['Chance of Admit'])
```

```python
plt.figure(figsize=(20,30),facecolor='white')
plotnumber = 1
for columns in X:
    if plotnumber<=15:
        ax = plt.subplot(5,3,plotnumber)
        plt.scatter(X[columns],y)
        plt.xlabel(columns,fontsize=20)
        plt.ylabel(columns,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
# here we are standardizing the data that means we are making sure that the mean = 0 and standard deviation = 1
```

Now we want to find the VIF(***Variance inflation factor***) because we want to find if there multicolinearity between the features

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled

vif = pd.DataFrame()

vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
vif['Features'] = X.columns
```

```python
vif
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.25,random_state=355)
```

```python
y_train
```

```python
lm = LinearRegression()

lm.fit(x_train,y_train)
```

```python
# save the model
filename = "final_model.pickle"
pickle.dump(lm,open(filename,'wb'))
```

```python
# prediction using the saved model
load_model = pickle.load(open(filename,'rb'))
a = load_model.predict(scaler.transform([[300,110,5,5,5,10,1]]))
a
```

```python
lm.score(x_train,y_train)
```

```python
adj_r2(x_train,y_train)
```

Now let's check the score on the test data so that we know that if there is any overfitting

```python
lm.score(x_test,y_test)
```

```python
# adjusted r2 for test data
adj_r2(x_test,y_test)
```

### Lasso Regularization

```python
lasso_model = LassoCV(alphas=None,cv=10,max_iter=100000)
lasso_model.fit(x_train,y_train)
```

```python
alpha = lasso_model.alpha_
alpha
```

```python
# now that we have the alpha value we will use it for regression model
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train,y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

### Ridge Regularization

```python
alphas = np.random.uniform(low=0,high=10,size=(50,))
ridgecv = RidgeCV(alphas=alphas,cv=10)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_reg = Ridge(alpha=ridgecv.alpha_)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

### Elastic Net

```python
elastic_cv = ElasticNetCV(alphas=None,cv=10)
elastic_cv.fit(x_train,y_train)
```

```python
elastic_cv.alpha_
```

```python
elastic_cv.l1_ratio_
```

```python
elasticnet_reg = ElasticNet(alpha=elastic_cv.alpha_,l1_ratio=0.5)
elasticnet_reg.fit(x_train,y_train)
```

```python
elasticnet_reg.score(x_test,y_test)
```
