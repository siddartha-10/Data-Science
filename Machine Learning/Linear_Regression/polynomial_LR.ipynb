```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
x = dataset.iloc[:,1:2].values
```

```python
x
```

```python
y = dataset.iloc[:,2].values
```

```python
y
```

```python
# visualization
plt.scatter(x,y,color='Red')
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x,y)
```

```python
plt.scatter(x,y,color='Red')
plt.plot(x,lm.predict(x),color='blue')
plt.title("Linear Regression")
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

```python
## we will try to fit the polynomial curve to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(x)
print(X_poly)
poly_reg.fit(X_poly,y)
```

```python
lm = LinearRegression()
lm.fit(X_poly,y)
```

```python
plt.scatter(x,y,color='Red')
plt.plot(x,lm.predict(poly_reg.fit_transform(x)),color='blue')
plt.title("Polynomial Regression")
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```

Even though we have the degree as 2 the line still doesn't fit well hence we will change the degree to 4

```python
poly_reg1 = PolynomialFeatures(degree=4)
X_poly1 = poly_reg1.fit_transform(x)
print(X_poly1)
poly_reg1.fit(X_poly1,y)
lm.fit(X_poly1,y)
```

```python
plt.scatter(x,y,color='Red')
plt.plot(x,lm.predict(X_poly1),color='blue')
plt.title("Polynomial Regression")
plt.xlabel("Levels")
plt.ylabel("Salary")
plt.show()
```
