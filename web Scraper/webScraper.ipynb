```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

## The below is used to make sure that the movies that you are getting are in english

```python
headers = {"Accept-Language": "en-US, en;q=0.5"}
```

## here i am trying to get the request the contents from the url

```python
url = "https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv"

results = requests.get(url, headers=headers)

## results store what the request.get, in this method the url will go the webpage and header is making sure that 
## we are only getting the names of them movies in english
```

## here i am trying to use beautifulSoup to make the resuklt more pretier

```python
soup = BeautifulSoup(results.text,"html.parser")

print(soup.prettify)
```
