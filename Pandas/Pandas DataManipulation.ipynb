## Read the cabin data and read the first letter of it and make a new column

```python
data_frame['Cabin']
```

```python
## converting the object into string using .astype(str)
char_cabin = data_frame["Cabin"].astype(str)
char_cabin
```

```python
## getting the first of each data row
new_cabin = [cabin[0] for cabin in char_cabin]
new_cabin
```

```python
new_cabin = pd.Categorical(new_cabin)
new_cabin
```

# Now add that new_cabin into the dataset with the column name as cabin_1

```python
data_frame['cabin_1'] = new_cabin
data_frame
```
