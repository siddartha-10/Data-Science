## Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

```python
pd.merge(transactions, users, on='UserID', how='left')
```

## Which transactions have a UserID not in users?

```python
missing = pd.merge(transactions, users, on='UserID', how='left', indicator=True)
```

```python
missing
## here i am using indicator hence i am able to get a new column called merge and this will contain right_only,left_only
## or both that implies if it exists in one table to 2 tables
```

```python
missing['_merge']
```

```python
missing_users = missing.loc[missing['_merge'] == 'left_only']
```

```python
missing_users
```

```python
pd.merge(transactions, users,on='UserID',how='left',indicator=True)
```

## Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

```python
pd.merge(transactions, users ,on='UserID')
```

## Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

```python
pd.merge(transactions, users,on='UserID',how='outer')
```

## Build a dataset with every possible (UserID, ProductID) pair (cross join)

```python
products.head()
```

```python
users.head()
```

```python
tuples = [(userID, productID) for userID in users['UserID'] for productID in products['ProductID']]
```

```python
tuples
```

```python
df_mul = pd.DataFrame(tuples,columns=['userID','ProductID'])
```

```python
df_mul
```

## Determine how much quantity of each product was purchased by each user

```python
transactions.head()
```

```python
df_qu = pd.merge(df_mul,pd.DataFrame(transactions['Quantity']),on=[df_mul['userID'],df_mul['ProductID']],how='left')
```
